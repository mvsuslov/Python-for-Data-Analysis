```python
# Загрузка библиотек
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
# считываем файл train.csv
data = pd.read_csv('/content/train.csv')
```

```python
# Survived
sns.countplot(x='Survived', data=data)
plt.title('Распределение Survived')
plt.xlabel('Выживаемость (0 = Нет, 1 = Да)')
plt.ylabel('Количество')
plt.show()

# Pclass
sns.countplot(x='Pclass', data=data)
plt.title('Распределение Pclass')
plt.xlabel('Класс пассажира')
plt.ylabel('Количество')
plt.show()

# Age
sns.histplot(data['Age'].dropna(), kde=True, bins=20)
plt.title('Распределение возраста')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()

# Sex
sns.countplot(x='Sex', data=data)
plt.title('Распределение по полу')
plt.xlabel('Пол')
plt.ylabel('Количество')
plt.show()

# Parch
sns.countplot(x='Parch', data=data)
plt.title('Распределение Parch')
plt.xlabel('Количество родителей/детей на борту')
plt.ylabel('Количество')
plt.show()
```

```python
sns.boxplot(x='Age', data=data)
plt.title('Boxplot для возраста')
plt.xlabel('Возраст')
plt.show()
```

```python
# Survived
survived_counts = data['Survived'].value_counts()
plt.pie(survived_counts, labels=['Не выжили', 'Выжили'], autopct='%1.1f%%', colors=['red', 'green'])
plt.title('Распределение выживаемости')
plt.show()

# Pclass
pclass_counts = data['Pclass'].value_counts()
plt.pie(pclass_counts, labels=['1 класс', '2 класс', '3 класс'], autopct='%1.1f%%', colors=['gold', 'silver', 'blue'])
plt.title('Распределение по классам')
plt.show()
```

```python
sns.pairplot(data.select_dtypes(include=['float64', 'int64']))
plt.suptitle('Pairplot для числовых переменных', y=1.02)
plt.show()
```

```python
fig = px.sunburst(data,
                  path=['Pclass', 'Sex'],
                  title='Sunburst: Пассажиры по классам и полу',
                  color='Pclass',
                  color_continuous_scale='RdBu')
fig.update_layout(title={'x': 0.5})
fig.show()
```
